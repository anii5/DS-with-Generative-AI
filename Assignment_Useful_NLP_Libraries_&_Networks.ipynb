1. Compare and contrast NLTK and spaCy in terms of features, ease of use, and performance.
  - When choosing between NLTK (Natural Language Toolkit) and spaCy for Natural Language Processing (NLP), the decision usually depends on whether you are conducting academic research or building a production-ready application.
      - 1. Features and Philosophy
          - NLTK (The Academic Library): NLTK is designed as a teaching and research tool. It provides a massive collection of algorithms for the same task (e.g., multiple different tokenizers or stemmers). This gives users maximum flexibility to experiment with different linguistic approaches. It also includes access to over 50 corpora and lexical resources like WordNet.
          - spaCy (The Industrial Library): spaCy is opinionated. Instead of giving you ten different ways to do one thing, it provides the single best-performing algorithm for a task. It focuses on providing a streamlined "pipeline" where text goes in and a rich doc object (containing tokens, entities, and syntax) comes out.
      - 2. Ease of Use
          - NLTK: NLTK can be more difficult for beginners because it requires manual string processing. You often have to write "glue code" to pass the output of one function (like a tokenizer) into another (like a POS tagger). Because it treats text primarily as strings or lists of strings, the workflow can feel fragmented.

          - spaCy: spaCy is designed for ease of use in a developer's workflow. It uses an object-oriented approach. When you process a string with spaCy, it automatically performs tokenization, tagging, and parsing in one step. The resulting Doc object allows you to access all linguistic features as attributes (e.g., token.pos_, token.dep_, doc.ents).
      - 3. Performance and Speed
          - NLTK: Because NLTK is written primarily in Python and focuses on modularity and experimentation, it is significantly slower. It is often used for processing smaller datasets or where execution time is not a critical factor.

          - spaCy: spaCy is built for speed. Its core is written in Cython, making it one of the fastest NLP libraries available. It is highly optimized for "information extraction" at scale and is the preferred choice for processing large volumes of real-world data in production environments.

4. Describe the architecture and functioning of a Recurrent Natural Network (RNN).
  - A Recurrent Neural Network (RNN) is a class of artificial neural networks designed to process sequential data. Unlike traditional feed-forward neural networks, which assume all inputs are independent of each other, RNNs have "memory." They take information from prior steps and use it to influence the current output.

This makes them the foundation for tasks like language translation, speech recognition, and time-series forecasting.

6. Write a Python program using TextBlob to perform sentiment analysis on the following paragraph of text:

“I had a great experience using the new mobile banking app. The interface is intuitive, and customer support was quick to resolve my issue. However, the app did crash once during a transaction, which was frustrating"

Your program should print out the polarity and subjectivity scores.

```python
pip install textblob
```

```python
from textblob import TextBlob

text = """I had a great experience using the new mobile banking app.
The interface is intuitive, and customer support was quick to resolve my issue.
However, the app did crash once during a transaction, which was frustrating."""

blob = TextBlob(text)

sentiment = blob.sentiment

print(f"Text: {text}\n")
print("-" * 30)
print(f"Polarity Score:     {sentiment.polarity:.4f}")
print(f"Subjectivity Score: {sentiment.subjectivity:.4f}")
print("-" * 30)

if sentiment.polarity > 0:
    print("Overall Sentiment: Positive")
elif sentiment.polarity < 0:
    print("Overall Sentiment: Negative")
else:
    print("Overall Sentiment: Neutral")
```

7. Given the sample paragraph below, perform string tokenization and frequency distribution using Python and NLTK:

“Natural Language Processing (NLP) is a fascinating field that combines linguistics, computer science, and artificial intelligence. It enables machines to understand, interpret, and generate human language. Applications of NLP include chatbots, sentiment analysis, and machine translation. As technology advances, the role of NLP in modern solutions is becoming increasingly critical.”

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

try:
    nltk.data.find('tokenizers/punkt_tab')
except LookupError:
    nltk.download('punkt_tab')

text = """Natural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence. It enables machines to understand, interpret,
and generate human language. Applications of NLP include chatbots, sentiment analysis,
and machine translation. As technology advances, the role of NLP in modern solutions
is becoming increasingly critical."""

tokens = word_tokenize(text)

clean_words = [w.lower() for w in tokens if w.isalnum()]

fdist = FreqDist(clean_words)

print("--- Processed Tokens (First 15) ---")
print(tokens[:15])

print("\n--- Word Frequency (Top 5) ---")
for word, count in fdist.most_common(5):
    print(f"{word:10} : {count}")

print("\n--- Summary ---")
print(f"Total tokens found: {len(tokens)}")
print(f"Unique words found: {len(fdist)}")
```

8.  Implement a basic LSTM model in Keras for a text classification task using the following dummy dataset. Your model should classify sentences as either positive (1) or negative (0).

Dataset texts = [ “I love this project”, #Positive “This is an amazing experience”, #Positive “I hate waiting in line”, #Negative “This is the worst service”, #Negative “Absolutely fantastic!” #Positive ] labels = [1, 1, 0, 0, 1]

Preprocess the text, tokenize it, pad sequences, and build an LSTM model to train on this data. You may use Keras with TensorFlow backend.  

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

texts = [
    "I love this project",
    "This is an amazing experience",
    "I hate waiting in line",
    "This is the worst service",
    "Absolutely fantastic!"
]
labels = np.array([1, 1, 0, 0, 1])
tokenizer = Tokenizer(num_words=100)
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)
max_length = max([len(x) for x in sequences])
padded_sequences = pad_sequences(sequences, maxlen=max_length, padding='post')
model = Sequential([
    Embedding(input_dim=100, output_dim=8, input_length=max_length),
    LSTM(16),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

print("Training the model...")
model.fit(padded_sequences, labels, epochs=10, verbose=0)

print("\n--- Model Summary ---")
model.summary()

print("\n--- Predictions on Training Data ---")
predictions = model.predict(padded_sequences)
for i, text in enumerate(texts):
    sentiment = "Positive" if predictions[i] > 0.5 else "Negative"
    print(f"Text: '{text}' -> Predicted: {sentiment} ({predictions[i][0]:.4f})")
```

9. Using spaCy, build a simple NLP pipeline that includes tokenization, lemmatization, and entity recognition. Use the following paragraph as your dataset:

“Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the development of India’s atomic energy program. He was the founding director of the Tata Institute of Fundamental Research (TIFR) and was instrumental in establishing the Atomic Energy Commission of India.”

Write a Python program that processes this text using spaCy, then prints tokens, their lemmas, and any named entities found.

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = ("Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role "
        "in the development of India’s atomic energy program. He was the founding "
        "director of the Tata Institute of Fundamental Research (TIFR) and was "
        "instrumental in establishing the Atomic Energy Commission of India.")

doc = nlp(text)

print("--- TOKENS & LEMMAS ---")

for token in doc:
    print(f"Token: {token.text:15} | Lemma: {token.lemma_}")

print("\n--- NAMED ENTITIES ---")
for ent in doc.ents:
    print(f"Entity: {ent.text:35} | Label: {ent.label_}")
```

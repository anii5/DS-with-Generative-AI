# Functions

1. What is the difference between a function and a method in Python?
  - **Function**
      - It is a block of organized, reusable code that performs a specific task.
      - It is defined independently, outside of any class.
      - It can be called by its name alone.
      - It operates on data passed to it as arguments.
      - It is used for general tasks or reusable blocks of code.
      - It may or may not return a value.
      - It is not associated with any specific object.
  - **Method**
      - It is also a block of code that performs a specific task.
      - It is defined within a class.
      - It is called on an object of the class.
      - It operates on the data associated with the object it is called on.
      - It encapsulates behavior associated with the object's data.
      - It always has the self parameter as its first argument, which refers to the instance of the class.
      - It can access and modify the object's attributes.
      - It is specific to object-oriented programming.

2. Explain the concept of function arguments and parameters in Python.
  - In Python, functions can accept inputs, which are specified using parameters and arguments.
      - Parameters
        - Parameters are the names listed in the function's definition, acting as placeholders for the values the function will receive.
        - They are essentially variables declared within the parentheses of the function definition.
        - Parameters define the type and number of inputs a function can accept, forming part of the function's signature.
      - Arguments
        - Arguments are the actual values passed to the function when it is called.
        - They provide the specific data that the function will operate on.
        - When a function is called, the arguments are matched to the parameters based on their position or keywords.

4. What is the purpose of the `return` statement in a Python function?
  - The return statement in Python serves two primary purposes within a function:
    - Exiting the function:
      - When a return statement is encountered, the function's execution immediately terminates, and control is passed back to the point where the function was called. Any code following the return statement within the function will not be executed.
    - Returning a value:
      - The return statement can optionally specify a value that the function will send back to the caller. This value can be of any data type, including numbers, strings, lists, dictionaries, or even other objects. If no value is specified, the function implicitly returns None.

5.  What are iterators in Python and how do they differ from iterables?
  - An iterable is an object that you can "iterate over," meaning you can go through its elements one by one. Many of the built-in data structures in Python are iterables.
      - Key characteristics of iterables:
        - Can be looped over: You can use them directly in a for loop.
        - Have an __iter__() method: When iter() is called on an iterable, it returns an iterator.

  - How they differ:
      - An iterable is a container of elements that can be iterated over.
      - An iterator is a stateful object that produces the next element in a sequence.
      - You get an iterator from an iterable by calling iter() on it.
      - Iterables must implement the __iter__() method.

6. Explain the concept of generators in Python and how they are defined.
  - Generators are a powerful and memory-efficient way to create iterators. They allow you to define a function that behaves like an iterator, producing a sequence of values on demand, one at a time, instead of building the entire sequence in memory. This "lazy evaluation" is their primary advantage.
      - Pausing execution: When a generator function encounters a yield statement, it pauses its execution, returns the value specified by yield, and saves its internal state.
      - Resuming execution: When the next value is requested (e.g., in a for loop or by calling next()), the generator resumes execution right from where it left off, continuing until it hits the next yield statement or finishes.
      - No StopIteration needed: Generators automatically handle the StopIteration exception when there are no more values to yield.

7. What are the advantages of using generators over regular functions?
 - A generator does not return its results all at once. Instead, it yields its values one by one, each time it is called. This makes it possible to generate an infinite sequence of values, as long as there is sufficient memory to store them.

It's typically used for short, simple operations and is often employed when a regular function definition would be too verbose or unnecessary, like when passing a function as an argument to another function.

9. Explain the purpose and usage of the `map()` function in Python.
  - The map() function in Python is a built-in higher-order function that applies a given function to each item of an iterable (like a list, tuple, or string) and returns a map object (an iterator) that yields the results. This function is a powerful tool for transforming data within iterables without the need for explicit loops, promoting cleaner and more concise code.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
  - map(), filter(), and reduce() are built-in functions in Python that facilitate functional programming paradigms. Here's a breakdown of their differences:
      - map()
          - Purpose: Applies a given function to each item of an iterable (e.g., list, tuple, etc.) and returns a new iterator with the results.
          - Output: Returns a map object (an iterator). You often need to convert it to a list, tuple, or other iterable to see the results.
          - Transformation: Transforms each element of the input iterable.
      - filter()
          - Purpose: Filters elements from an iterable based on a given function (predicate). It keeps only the elements for which the predicate returns True.
          - Output: Returns a filter object (an iterator). You need to convert it to a list, tuple, or other iterable to see the results.
          - Selection: Selects elements based on a condition.
      - reduce()
          - Purpose: Applies a function cumulatively to the items of an iterable, reducing it to a single value.
          - Output: Returns a single value.
          - Aggregation: Aggregates elements into a single result.
          - Import: Requires importing from the functools module.

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_even_numbers(numbers):
    even_sum = 0
    for number in numbers:
        if number % 2 == 0:
            even_sum += number
    return even_sum
list = [1,8,4,6,9,44,15,32]
print(f"List: {list}")
print(f"Sum of even numbers: {sum_even_numbers(list)}")
```

2. Create a Python function that accepts a string and returns the reverse of that string.

```python
def reverse_string(s: str) -> str:
  return s[::-1]
original_string = "Aniket Apurv"
reversed_str = reverse_string(original_string)
print(f"Original string: {original_string}")
print(f"Reversed string: {reversed_str}")
```

3.  Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

```python
def square_list(numbers):
    return [x ** 2 for x in numbers]
input_list = [1, 2, 3, 4, 5,6]
squared_list = square_list(input_list)
print(squared_list)
```

4.  Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
is_prime = lambda num: num > 1 and all(num % i for i in range(2, int(num/2) + 1))

for number in range(1, 201):
    if is_prime(number):
        print(f"{number} is a prime number")
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i
for power in powers_of_2(5):
    print(power)
```

7.  Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line

file_path = 'example.txt'
with open(file_path, 'w') as f:
    f.write("My Name is Aniket Apurv\n")
    f.write("I have completed B.Sc CS\n")
    f.write("Now studying DS")
for line in read_file_line_by_line(file_path):
    print(line, end='')
```

8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

```python
my_list = [('a', 3), ('b', 1), ('c', 2)]
sorted_list = sorted(my_list, key=lambda x: x[1])
print(sorted_list)
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
def celsius_to_fahrenheit(celsius):
  return (celsius * 9/5) + 32
def convert_temperatures(celsius_list):
  fahrenheit_list = list(map(celsius_to_fahrenheit, celsius_list))
  return fahrenheit_list
celsius_temps = [0, 10, 20, 30, 40]
fahrenheit_temps = convert_temperatures(celsius_temps)
print(f"Celsius temperatures: {celsius_temps}")
print(f"Fahrenheit temperatures: {fahrenheit_temps}")
```

```python
a=list("Aniket Apurv")
def vowel(x):
    v=('a','e','i','o','u')
    return x in v

b=list(filter(vowel,a))
print("Vowels : ",b)
```

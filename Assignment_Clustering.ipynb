#Theoretical

1. What is unsupervised learning in the context of machine learning?
  - Unsupervised learning is a type of machine learning where algorithms analyze unlabeled data to find hidden patterns, structures, and relationships without human intervention. It differs from supervised learning because it doesn't use a "teacher" or pre-defined correct answers to guide the model. Instead, the algorithm learns by discovering underlying patterns, grouping similar data points together through methods like clustering, and reducing data complexity.  

2. How does K-Means clustering algorithm work?
  - K-means clustering works by iteratively assigning data points to the closest of K clusters and then recalculating the cluster centroids until the assignments no longer change. The process involves selecting the number of clusters (K), randomly placing initial centroids, and then repeating the steps of assigning each data point to its nearest centroid and updating each centroid to the mean of its assigned points. This continues until the centroids stabilize, which minimizes the total distance between each point and its cluster's centroid.  

3. Explain the concept of a dendrogram in hierarchical clustering.
  - A dendrogram is a tree-like diagram used in hierarchical clustering to visualize the relationships between data points. It illustrates how clusters are merged, step-by-step, with the height of the vertical lines indicating the distance or dissimilarity between merged clusters. By cutting the dendrogram horizontally at a certain height, you can determine the number of clusters at that level of dissimilarity.  

4. What is the main difference between K-Means and Hierarchical Clustering?
  - The main difference is that k-means requires you to specify the number of clusters (\(k\)) beforehand, while hierarchical clustering does not, instead building a hierarchy of clusters visualized as a dendrogram. K-means partitions data into a fixed number of non-overlapping clusters, whereas hierarchical clustering builds nested clusters by either merging or splitting them, creating a tree-like structure that reveals the relationships between clusters at different levels.

5. What are the advantages of DBSCAN over K-Means?
  - DBSCAN is advantageous over K-Means because it does not require a pre-specified number of clusters, can find arbitrarily shaped clusters, and effectively identifies outliers. In contrast, K-Means assumes clusters are spherical, requires the number of clusters to be set beforehand, and can struggle with non-globular data. DBSCAN is more resistant to noise and can find clusters of varying densities, which is not possible with K-Means.

6. When would you use Silhouette Score in clustering?
  - You would use Silhouette Score to evaluate the quality of a clustering result by measuring how similar each data point is to its own cluster compared to other clusters. It is used to determine the optimal number of clusters (k) for an algorithm and to validate the clustering performance by providing a single score for the entire dataset.

7. What are the limitations of Hierarchical Clustering?
  - Hierarchical clustering's main limitations include its high computational cost and memory usage, making it difficult to scale to large datasets. It is also sensitive to noise and outliers, and once a merge or split is made, the decision is irreversible, which can lead to suboptimal clusters.

8. Why is feature scaling important in clustering algorithms like K-Means?
  - Feature scaling is important for K-Means because it's a distance-based algorithm, and features with larger ranges would disproportionately influence the distance calculations, skewing the clusters. Scaling ensures all features contribute equally to determining similarity, preventing features with high values from dominating the analysis and improving clustering accuracy.

9. How does DBSCAN identify noise points?
  - DBSCAN identifies noise points by first classifying data points based on density and then labeling any point that doesn't meet the criteria for a cluster as noise. A point is labeled as noise if it is not within the specified distance (\(\varepsilon \)) of a sufficient number of other points (defined by MinPts) to be considered a core point. These isolated points are assigned a noise label of -1 and are not included in any cluster.

10.  Define inertia in the context of K-Means.
  - In K-Means, inertia is the sum of the squared distances between each data point and the centroid of its assigned cluster. It is a measure of the internal coherence or compactness of the clusters. A lower inertia score indicates that data points are closer to their cluster centers, meaning the clusters are more dense and cohesive.

11. What is the elbow method in K-Means clustering?
  - The elbow method is a heuristic technique for determining the optimal number of clusters, k, in a dataset for K-Means clustering. It involves plotting the Within-Cluster Sum of Squares (WCSS) against a range of k values and identifying the "elbow" point on the graph, which is where the rate of decrease in WCSS sharply slows down. This "elbow" represents a point of diminishing returns, indicating that adding more clusters past this point provides little improvement in reducing the total within-cluster variance.  

12. Describe the concept of "density" in DBSCAN.
  - In DBSCAN, "density" refers to the concept of how closely packed data points are in a given area. The algorithm uses two main parameters, epsilon (\(\epsilon \)) (a radius) and minPoints (the minimum number of points), to define what constitutes a dense region. Clusters are identified as continuous areas of high density, while points in low-density regions are labeled as noise or outliers.

13.  Can hierarchical clustering be used on categorical data?
  - Yes, hierarchical clustering can be used on categorical data by defining a specific dissimilarity measure, such as Hamming distance, instead of the traditional Euclidean distance used for numerical data. This requires creating a dissimilarity matrix based on how different the categories are for each data point.
  

14. What does a negative Silhouette Score indicate?
  - A negative silhouette score indicates that the data points are likely assigned to the wrong cluster. This means that the points are, on average, more similar to the data points in a different cluster than they are to the data points in their own cluster. In other words, the clustering algorithm has failed to correctly group the data, and the clusters are poorly separated.

15. Explain the term "linkage criteria" in hierarchical clustering.
  - Linkage criteria in hierarchical clustering define how the distance between two clusters is calculated, which determines how they are merged during the clustering process. Common criteria include single linkage (minimum distance between any two points), complete linkage (maximum distance between any two points), average linkage (average of all pairwise distances), and Ward's method (merging clusters to minimize the increase in total within-cluster variance). The choice of linkage method significantly impacts the shape and size of the resulting clusters.

16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?
  - K-Means clustering may perform poorly on data with varying cluster sizes or densities due to its inherent assumptions and methodology.
      - Varying Cluster Sizes: If clusters have significantly different sizes, K-Means may struggle to accurately identify the smaller clusters. Larger clusters, with more data points, exert a stronger pull on their centroids, potentially drawing data points from smaller, nearby clusters and causing misclassification.
      - Varying Cluster Densities: When clusters have different densities, K-Means may incorrectly group data points. A dense cluster might be inappropriately split into multiple smaller, less dense "clusters" if its overall size is large, while a sparse, but coherent, cluster might be absorbed into a neighboring, denser cluster due to the centroid's pull.
      - Sensitivity to Outliers: K-Means is sensitive to outliers, especially when dealing with varying densities. Outliers can significantly shift the mean (centroid) of a cluster, leading to distorted cluster boundaries and misclassification of other data points, particularly in sparser regions.

17. What are the core parameters in DBSCAN, and how do they influence clustering?
  - The core parameters in the DBSCAN (Density-Based Spatial Clustering of Applications with Noise) algorithm are epsilon (Îµ) and minimum points (MinPts). By defining a neighborhood around data points, these parameters allow DBSCAN to identify densely packed regions as clusters and separate them from noise or outliers.
  
      - \(Epsilon(\varepsilon )\) (or eps): This parameter defines the maximum radius of the neighborhood around each data point. It is a distance measure used to find neighboring points. If the distance between two points is less than or equal to \(\varepsilon \), they are considered neighbors.
      - \(Minimum\ points(MinPts)\) (or min_samples): This parameter sets the minimum number of data points required within an \(\varepsilon \) radius to form a dense region. A point is classified as a "core point" if it has at least MinPts neighbors inside its \(\varepsilon \)-neighborhood.

18.  How does K-Means++ improve upon standard K-Means initialization?
  - K-Means++ improves upon standard K-Means by using a smarter, probabilistic initialization that spreads out the initial centroids, leading to better and faster convergence. Unlike standard K-Means, which picks initial centroids randomly, K-Means++ selects the first centroid uniformly at random and then chooses subsequent centroids based on a probability proportional to their squared distance from the nearest existing centroid. This approach ensures initial centroids are far apart, which increases the likelihood of finding a better final solution and avoids poor clusterings where multiple centroids end up in the same cluster.

19. What is agglomerative clustering?
  - Agglomerative clustering is a "bottom-up" hierarchical clustering method that starts with each data point as its own cluster and iteratively merges the closest pairs of clusters until only one cluster remains. It creates a hierarchy of clusters, which can be visualized as a dendrogram, and does not require the number of clusters to be specified beforehand. Key steps include calculating a distance matrix, merging the closest clusters based on a linkage criterion, and updating the distance matrix until the process is complete.  

20. What makes Silhouette Score a better metric than just inertia for model evaluation?
  - Silhouette Score is a better metric than inertia because it evaluates both the cohesion (how close data points are within a cluster) and the separation (how far apart clusters are from each other), while inertia only measures compactness and lacks context about cluster separation. A high inertia could result from having too many small clusters, whereas Silhouette Score provides a more comprehensive measure of a clustering solution's overall quality.

21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a
scatter plot

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

n_samples = 300
n_features = 2
n_centers = 4
random_state = 42

X, y_true = make_blobs(n_samples=n_samples, n_features=n_features,
                       centers=n_centers, cluster_std=0.8, random_state=random_state)
kmeans = KMeans(n_clusters=n_centers, random_state=random_state, n_init=10)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
centers = kmeans.cluster_centers_
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis', label='Data Points')
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200, alpha=0.7, label='Cluster Centroids')
plt.title('K-Means Clustering with 4 Centers')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```

23. Generate synthetic data using make_moons and apply DBSCAN. Highlight outliers in the plot

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import numpy as np
X, y = make_moons(n_samples=200, noise=0.05, random_state=42)
dbscan = DBSCAN(eps=0.3, min_samples=5)
clusters = dbscan.fit_predict(X)
plt.figure(figsize=(8, 6))
outliers = X[clusters == -1]
unique_clusters = np.unique(clusters)
for cluster_label in unique_clusters:
    if cluster_label == -1:
        continue
    cluster_points = X[clusters == cluster_label]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {cluster_label}', s=50, alpha=0.7)
plt.scatter(outliers[:, 0], outliers[:, 1], color='red', marker='x', s=100, label='Outliers')

plt.title('DBSCAN Clustering with Outliers Highlighted (make_moons data)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```

24.  Load the Wine dataset and apply K-Means clustering after standardizing the features. Print the size of each

26.  Load the Breast Cancer dataset, apply MinMaxScaler, and use K-Means with 2 clusters. Output the cluster
centroids

```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
import pandas as pd
data = load_breast_cancer(as_frame=True)
X = data.data
feature_names = data.feature_names
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=feature_names)
kmeans = KMeans(n_clusters=2, random_state=42, n_init='auto')
kmeans.fit(X_scaled_df)
centroids_scaled = kmeans.cluster_centers_
centroids_df = pd.DataFrame(centroids_scaled, columns=feature_names, index=[f'Cluster {i}' for i in range(kmeans.n_clusters)])
centroids_original = scaler.inverse_transform(centroids_scaled)
centroids_original_df = pd.DataFrame(centroids_original, columns=feature_names, index=[f'Cluster {i}' for i in range(kmeans.n_clusters)])

print("Centroids (in Scaled [0, 1] Domain):")
print(centroids_df)

print("\n" + "="*50 + "\n")

print("Centroids (in Original Feature Domain for Interpretation):")
print(centroids_original_df)
```

27. Generate synthetic data using make_blobs with varying cluster standard deviations and cluster with
DBSCAN

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

n_samples = 1500
n_features = 2
centers = 4
cluster_stds = [0.6, 0.3, 0.9, 0.4]
random_state = 42

X, y_true = make_blobs(n_samples=n_samples, n_features=n_features,
                       centers=centers, cluster_std=cluster_stds,
                       random_state=random_state)
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y_true, cmap='viridis', s=50, alpha=0.7)
plt.title('True Clusters (Varying Std Dev)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.subplot(1, 2, 2)
unique_labels = set(dbscan_labels)
colors = plt.cm.viridis(np.linspace(0, 1, len(unique_labels) - (1 if -1 in unique_labels else 0)))

for k, col in zip(unique_labels, colors):
    if k == -1:
        plt.scatter(X[dbscan_labels == k, 0], X[dbscan_labels == k, 1],
                    c='black', marker='x', s=50, label='Noise')
    else:
        plt.scatter(X[dbscan_labels == k, 0], X[dbscan_labels == k, 1],
                    c=[col], s=50, alpha=0.7, label=f'Cluster {k}')

plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.tight_layout()
plt.show()
```

28.  Load the Digits dataset, reduce it to 2D using PCA, and visualize clusters from K-Means

29.  Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score

X, y = make_blobs(
    n_samples=500,
    n_features=2,
    centers=4,  #
    cluster_std=1.0,
    random_state=42,
)

silhouette_scores = []
k_values = range(2, 6)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    score = silhouette_score(X, kmeans.labels_)
    silhouette_scores.append(score)
    print(f"Silhouette Score for k={k}: {score:.4f}")
plt.figure(figsize=(8, 6))
plt.bar(
    [str(k) for k in k_values],
    silhouette_scores,
    color="skyblue",
)
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for Different Numbers of Clusters")
plt.ylim(0, 1)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

30.  Load the Iris dataset and use hierarchical clustering to group data. Plot a dendrogram with average linkage

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import linkage, dendrogram

iris = load_iris()
X = iris.data

linked = linkage(X, method='average')

plt.figure(figsize=(10, 7))
dendrogram(linked,
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram (Average Linkage)')
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distance')
plt.show()
```

31.  Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with
decision boundaries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

n_samples = 500
n_features = 2
n_clusters = 3
cluster_std = 1.5
random_state = 42

X, y = make_blobs(n_samples=n_samples, n_features=n_features, centers=n_clusters,
                  cluster_std=cluster_std, random_state=random_state)

kmeans = KMeans(n_clusters=n_clusters, random_state=random_state, n_init=10)
kmeans.fit(X)
y_pred = kmeans.predict(X)
centroids = kmeans.cluster_centers_

plt.figure(figsize=(10, 7))

plt.scatter(X[:, 0], X[:, 1], c=y_pred, s=50, cmap='viridis', alpha=0.7, label='Data Points')

plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, alpha=0.9, marker='X', label='Centroids')

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')

plt.title('K-Means Clustering with Overlapping Clusters and Decision Boundaries')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```

32. Load the Digits dataset and apply DBSCAN after reducing dimensions with t-SNE. Visualize the results

33. Generate synthetic data using make_blobs and apply Agglomerative Clustering with complete linkage. Plot
the result

34.  Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a
line plot

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler

cancer_data = load_breast_cancer()
X = cancer_data.data

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

inertia = []

k_values = range(2, 7)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')

    kmeans.fit(X_scaled)

    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(k_values, inertia, marker='o')
plt.title('Inertia vs. Number of Clusters (K)')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.xticks(k_values)
plt.grid(True)
plt.show()

print("Inertia values for K=2 to 6:")
for k, val in zip(k_values, inertia):
    print(f"K = {k}: {val:.2f}")
```

36.  Use the Wine dataset, apply DBSCAN after scaling the data, and count the number of clusters (excluding

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors

wine = datasets.load_wine()
X = wine.data
y = wine.target

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

dbscan = DBSCAN(eps=0.8, min_samples=25)
clusters = dbscan.fit_predict(X_scaled)

n_clusters = len(set(clusters)) - (1 if -1 in clusters else 0)

n_noise = np.sum(clusters == -1)

print(f"Estimated number of clusters: {n_clusters}")
print(f"Number of noise points: {n_noise}")

from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(10, 7))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='viridis')
plt.title('DBSCAN Clustering of Wine Dataset (after PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(handles=scatter.legend_elements()[0], title='Clusters')
plt.show()
```

44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.metrics import silhouette_samples, silhouette_score

iris = load_iris()
X = iris.data

n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(X)

silhouette_per_sample = silhouette_samples(X, cluster_labels)

plt.figure(figsize=(10, 6))
y_lower = 10

average_silhouette_score = silhouette_score(X, cluster_labels)
print(f"Average Silhouette Score: {average_silhouette_score:.2f}")


for i in range(n_clusters):

    ith_cluster_silhouette_values = silhouette_per_sample[cluster_labels == i]
    ith_cluster_silhouette_values.sort()

    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i

    color = plt.cm.nipy_spectral(float(i) / n_clusters)
    plt.fill_betweenx(
        np.arange(y_lower, y_upper),
        0,
        ith_cluster_silhouette_values,
        facecolor=color,
        edgecolor=color,
        alpha=0.7,
    )

    plt.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

    y_lower = y_upper + 10

plt.title("Silhouette plot for the various clusters")
plt.xlabel("The Silhouette Coefficient values")
plt.ylabel("Cluster label")

plt.axvline(x=average_silhouette_score, color="red", linestyle="--")

plt.yticks([])
plt.xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
plt.show()
```

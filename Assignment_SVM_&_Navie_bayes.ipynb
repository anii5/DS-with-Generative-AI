1. What is a Support Vector Machine (SVM)?
  - A Support Vector Machine (SVM) is a supervised machine learning algorithm that classifies data by finding an optimal "hyperplane" to separate different classes. It works by finding the hyperplane that has the largest margin, or distance, to the nearest data points of each class. These points, called support vectors, are the most critical for defining the separation boundary. SVMs are used for both classification and regression, and they can handle both linear and nonlinear data by using kernel functions to transform the data into a higher-dimensional space where it becomes linearly separable.  

2.  What is the difference between Hard Margin and Soft Margin SVM?
  - Hard margin SVM requires perfect linear separation and cannot tolerate misclassified points or outliers, which can cause it to be sensitive to noise and lead to overfitting. In contrast, soft margin SVM allows for some misclassifications by introducing "slack variables," which trade off margin maximization with error minimization to achieve better generalization and robustness against outliers.

3. What is the mathematical intuition behind SVM?
  - The mathematical intuition behind a Support Vector Machine (SVM) revolves around three core concepts: maximizing the margin, leveraging "support vectors" to define the decision boundary, and using the "kernel trick" to handle non-linear data. This approach makes SVM a robust and powerful classifier.

4. What is the role of Lagrange Multipliers in SVM?
  - Lagrange multipliers transform the SVM's constrained optimization problem into an unconstrained one, enabling a solution through its "dual formulation". This method allows the algorithm to focus on the support vectors, which are the data points with non-zero Lagrange multipliers, simplifying the optimization process and reducing computational complexity.

5. What are Support Vectors in SVM?
  - Support vectors in Support Vector Machines (SVM) are the data points closest to the decision boundary, which define the hyperplane and are crucial for classification. These vectors are called support vectors because they "support" the hyperplane; they are the critical data points that determine the position of the decision boundary and the size of the margin between classes. The SVM's ability to generalize well and classify new data accurately relies on these support vectors.

6.  What is a Support Vector Classifier (SVC)?
  - A Support Vector Classifier (SVC) is a supervised machine learning model that finds an optimal hyperplane to separate data points into two classes. It works by maximizing the margin, or distance, between the hyperplane and the nearest data points, known as support vectors. SVCs are a type of Support Vector Machine (SVM), specifically used for classification, and they can handle non-linearly separable data by mapping it to a higher-dimensional space, allowing for more complex decision boundaries

7. What is a Support Vector Regressor (SVR)?
  - Support Vector Regression (SVR) is a supervised machine learning technique that uses Support Vector Machines (SVMs) for regression tasks to find a function that fits data by minimizing prediction error. SVR creates an epsilon-insensitive tube around the predicted function, ignoring points within the tube and penalizing those outside, to maximize the margin of error tolerance and find the best-fitting hyperplane. It handles non-linear data by mapping it to a higher-dimensional space using kernel functions.

8. What is the Kernel Trick in SVM?
  - The kernel trick is a technique used in Support Vector Machines (SVM) to handle non-linearly separable data by implicitly mapping the data into a higher-dimensional space, where it becomes linearly separable. Instead of explicitly computing the coordinates of data points in this new high-dimensional space, a kernel function calculates their dot products directly. This allows the SVM to find a linear decision boundary in that higher dimension without the high computational cost of the explicit transformation, making complex, non-linear classification problems solvable.

10. What is the effect of the C parameter in SVM?
  - The C parameter in Support Vector Machines (SVMs) controls the trade-off between maximizing the margin and minimizing misclassifications. A high C value penalizes misclassified training points heavily, leading to a narrower margin but potentially better classification of training data. Conversely, a low C value allows for more misclassified points, resulting in a wider margin and a more generalized model that may perform better on unseen data.

11.  What is the role of the Gamma parameter in RBF Kernel SVM?
  - Gamma controls the "reach" or influence of a single training example in an RBF kernel SVM. A high gamma value means each training point has a very local influence, leading to a more complex, "wiggly" decision boundary that can overfit the data. Conversely, a low gamma value gives each point a wide influence, resulting in a smoother, less complex boundary that might underfit the data.  

12. What is the Naïve Bayes classifier, and why is it called "Naïve"?
  - The Naïve Bayes classifier is a probabilistic machine learning algorithm based on Bayes' Theorem that uses the "naïve" assumption that all features are independent of one another. It is called "naïve" because this assumption is often unrealistic in real-world data, as features are frequently correlated. Despite this simplification, the classifier is highly effective and widely used for tasks like spam filtering and text classification.

13.  What is Bayes’ Theorem?
  - Bayes' theorem is a mathematical formula used to update the probability of a hypothesis based on new evidence. It connects the prior probability of an event with new information (likelihood) to calculate a more accurate probability, called the posterior probability. In essence, it provides a way to revise existing beliefs in light of new data.  

14.  Explain the differences between Gaussian Naïve Bayes, Multinomial Naïve Bayes, and Bernoulli Naïve Bayes.
  - The key difference between Gaussian Naïve Bayes, Multinomial Naïve Bayes, and Bernoulli Naïve Bayes lies in how they handle the distribution of features, with Gaussian assuming continuous data, Multinomial handling discrete counts, and Bernoulli focusing on binary presence/absence of features.

15. When should you use Gaussian Naïve Bayes over other variants?
  - when your data has continuous numerical features that can be assumed to follow a normal (Gaussian) distribution. It is best for problems where features like height, weight, or temperature are used for classification, as it calculates the mean and standard deviation for each feature within each class to estimate the probability density function.  

16. What are the key assumptions made by Naïve Bayes?
  - The core assumption of Naive Bayes is feature independence, meaning it assumes that all predictive features (or attributes) are independent of each other, given the class label. This "naive" assumption simplifies complex probability calculations by allowing the algorithm to treat each feature's contribution to the overall probability as separate, even though this independence rarely holds true in real-world scenarios.

17.  What are the advantages and disadvantages of Naïve Bayes?
  - Naïve Bayes is advantageous for its simplicity, speed, and scalability, making it easy to implement and fast to train, especially with large datasets. However, its main disadvantage is the naïve assumption of feature independence, which can lead to poor performance if the assumption is violated. Another drawback is the "zero-frequency problem," where a feature not present in the training data can lead to a zero probability, though this is often handled with smoothing techniques.

18. Why is Naïve Bayes a good choice for text classification?
  - Naïve Bayes is a good choice for text classification because it is computationally efficient, fast to train, and handles high-dimensional data effectively. It is also simple to implement and works well even with small datasets. While its core assumption of word independence is often not true in reality, it has been shown to perform well in practice for tasks like spam filtering, sentiment analysis, and news classification.

19.  Compare SVM and Naïve Bayes for classification tasks.
  - SVM finds an optimal hyperplane to separate data, while Naïve Bayes is a probabilistic classifier that assumes feature independence and predicts the class with the highest probability. SVMs generally achieve higher accuracy and are more robust, especially with larger datasets, whereas Naïve Bayes is computationally efficient, scales well with large datasets, and is simple to implement. The best choice depends on the specific task, with SVM being better for complex scenarios and Naïve Bayes for quick, large-scale classification.

20.  How does Laplace Smoothing help in Naïve Bayes?
  - Laplace smoothing helps Naïve Bayes by preventing zero probabilities by adding a small value (like 1) to the counts of all features, ensuring that no feature is completely absent in the training data. This is crucial because a zero probability for any feature would make the entire probability calculation zero, leading to inaccurate predictions and misclassification errors, especially with small datasets. It makes the model more robust and better able to generalize to new, unseen data.  

21. Write a Python program to train an SVM Classifier on the Iris dataset and evaluate accuracy.

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

def train_and_evaluate_svm():
    print("--- Starting SVM Classification on Iris Dataset ---")

    iris = load_iris()
    X = iris.data
    y = iris.target

    print(f"Total samples loaded: {len(X)}")
    print(f"Number of features: {X.shape[1]}")
    print(f"Target classes: {iris.target_names}")

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42, stratify=y
    )

    print(f"\nTraining set size: {len(X_train)} samples")
    print(f"Testing set size: {len(X_test)} samples")

    svm_classifier = SVC(kernel='linear', C=1.0, random_state=42)
    print("\nTraining an SVM Classifier with a Linear Kernel...")

    svm_classifier.fit(X_train, y_train)

    y_pred = svm_classifier.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)

    print("\n--- Model Evaluation Results ---")
    print(f"Accuracy: {accuracy:.4f} (proportion of correctly classified samples)")

    report = classification_report(y_test, y_pred, target_names=iris.target_names)
    print("\nClassification Report (Precision, Recall, F1-Score per class):\n")
    print(report)

if __name__ == "__main__":
    np.random.seed(42)
    train_and_evaluate_svm()
```

22.  Write a Python program to train two SVM classifiers with Linear and RBF kernels on the Wine dataset, then
compare their accuracies.

```python
import numpy as np
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import time

def compare_svm_kernels_on_wine():

    print("--- Starting SVM Kernel Comparison on Wine Dataset ---")

    wine = load_wine()
    X = wine.data
    y = wine.target

    print(f"Total samples loaded: {len(X)}")
    print(f"Number of features: {X.shape[1]}")
    print(f"Target classes: {wine.target_names}")

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42, stratify=y
    )

    print(f"\nTraining set size: {len(X_train)} samples")
    print(f"Testing set size: {len(X_test)} samples")

    results = {}

    svm_linear = SVC(kernel='linear', C=1.0, random_state=42)
    print("\nTraining SVM with Linear Kernel...")

    start_time_linear = time.time()
    svm_linear.fit(X_train, y_train)
    end_time_linear = time.time()

    y_pred_linear = svm_linear.predict(X_test)
    accuracy_linear = accuracy_score(y_test, y_pred_linear)

    results['Linear'] = {
        'accuracy': accuracy_linear,
        'time': end_time_linear - start_time_linear
    }

    svm_rbf = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42)
    print("Training SVM with RBF (Gaussian) Kernel...")

    start_time_rbf = time.time()
    svm_rbf.fit(X_train, y_train)
    end_time_rbf = time.time()

    y_pred_rbf = svm_rbf.predict(X_test)
    accuracy_rbf = accuracy_score(y_test, y_pred_rbf)

    results['RBF'] = {
        'accuracy': accuracy_rbf,
        'time': end_time_rbf - start_time_rbf
    }

    print("\n\n===================================")
    print("      Kernel Performance Comparison")
    print("===================================")

    print(f"Linear Kernel Accuracy: {results['Linear']['accuracy']:.4f}")
    print(f"RBF Kernel Accuracy:    {results['RBF']['accuracy']:.4f}")

    print("\n--- Training Time Comparison ---")
    print(f"Linear Kernel Training Time: {results['Linear']['time']:.5f} seconds")
    print(f"RBF Kernel Training Time:    {results['RBF']['time']:.5f} seconds")

    print("\n--- Conclusion ---")
    if accuracy_linear > accuracy_rbf:
        print("The Linear Kernel achieved higher accuracy.")
    elif accuracy_rbf > accuracy_linear:
        print("The RBF Kernel achieved higher accuracy.")
    else:
        print("Both kernels achieved the same accuracy.")


if __name__ == "__main__":
    np.random.seed(42)
    compare_svm_kernels_on_wine()
```

23. Write a Python program to train an SVM Regressor (SVR) on a housing dataset and evaluate it using Mean
Squared Error (MSE)

```python
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
import time

def train_and_evaluate_svr():

    print("--- Starting SVR Regression on California Housing Dataset ---")

    housing = fetch_california_housing(as_frame=False)
    X = housing.data
    y = housing.target

    print(f"Total samples loaded: {len(X)}")
    print(f"Number of features: {X.shape[1]}")

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42
    )

    print(f"\nTraining set size: {len(X_train)} samples")
    print(f"Testing set size: {len(X_test)} samples")

    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    svr_regressor = SVR(kernel='rbf', C=10.0, epsilon=0.1)
    print("\nTraining SVR Regressor with RBF Kernel (C=10.0, Epsilon=0.1)...")

    start_time = time.time()
    svr_regressor.fit(X_train_scaled, y_train)
    end_time = time.time()

    y_pred = svr_regressor.predict(X_test_scaled)

    mse = mean_squared_error(y_test, y_pred)

    r2 = r2_score(y_test, y_pred)

    print("\n--- Model Evaluation Results ---")
    print(f"Training Time: {end_time - start_time:.4f} seconds")
    print(f"Mean Squared Error (MSE): {mse:.4f}")
    print(f"R-squared (R2) Score:     {r2:.4f}")

    print("\nNote: MSE represents the average squared difference between the predicted and actual values.")

if __name__ == "__main__":
    np.random.seed(42)
    train_and_evaluate_svr()
```

24.  Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision
boundary.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_circles # Good for demonstrating non-linear boundaries

def plot_decision_boundary(X, y, classifier, title="SVM Decision Boundary"):

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                         np.arange(y_min, y_max, 0.02))

    Z = classifier.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(10, 8))

    plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)

    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, s=50, edgecolors='k')

    plt.scatter(classifier.support_vectors_[:, 0],
                classifier.support_vectors_[:, 1],
                s=150, facecolors='none', edgecolors='green', label='Support Vectors', linewidth=2)

    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

def train_poly_svm_and_visualize():
    print("--- Starting Polynomial SVM Decision Boundary Visualization ---")

    X, y = make_circles(n_samples=100, factor=0.5, noise=0.05, random_state=42)

    print(f"Generated {len(X)} samples with {X.shape[1]} features.")
    print(f"Number of samples in class 0: {np.sum(y == 0)}")
    print(f"Number of samples in class 1: {np.sum(y == 1)}")


    poly_svm_classifier = svm.SVC(kernel='poly', degree=3, C=1.0, gamma='scale', coef0=1, random_state=42)
    print("\nTraining SVM Classifier with Polynomial Kernel (degree=3, C=1.0, coef0=1)...")

    poly_svm_classifier.fit(X, y)

    plot_decision_boundary(X, y, poly_svm_classifier,
                           title="Polynomial SVM Decision Boundary (Degree=3)")
    print("\nVisualization complete. A plot window should have appeared.")

if __name__ == "__main__":
    np.random.seed(42)
    train_poly_svm_and_visualize()
```

25. Write a Python program to train a Gaussian Naïve Bayes classifier on the Breast Cancer dataset and
evaluate accuracy

```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler

def train_and_evaluate_gaussian_nb():
    print("--- Starting Gaussian Naïve Bayes on Breast Cancer Dataset ---")

    cancer = load_breast_cancer()
    X = cancer.data
    y = cancer.target

    print(f"Total samples loaded: {len(X)}")
    print(f"Number of features: {X.shape[1]}")
    print(f"Target classes: {cancer.target_names}")

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42, stratify=y
    )

    print(f"\nTraining set size: {len(X_train)} samples")
    print(f"Testing set size: {len(X_test)} samples")

    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    nb_classifier = GaussianNB()
    print("\nTraining Gaussian Naïve Bayes Classifier...")

    nb_classifier.fit(X_train_scaled, y_train)

    y_pred = nb_classifier.predict(X_test_scaled)

    accuracy = accuracy_score(y_test, y_pred)

    print("\n--- Model Evaluation Results ---")
    print(f"Accuracy Score: {accuracy:.4f}")

    print("\nClassification Report:")
    print(classification_report(y_test, y_pred, target_names=cancer.target_names))

if __name__ == "__main__":
    np.random.seed(42)
    train_and_evaluate_gaussian_nb()
```

26. Write a Python program to train a Multinomial Naïve Bayes classifier for text classification using the 20
Newsgroups dataset

```python
import numpy as np
from sklearn.datasets import fetch_20newsgroups
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
import time

def train_and_evaluate_multinomial_nb():

    print("--- Starting Multinomial Naïve Bayes Text Classification ---")

    categories = ['alt.atheism', 'soc.religion.christian',
                  'comp.graphics', 'sci.med']

    print(f"Loading data for categories: {categories}...")

    newsgroups_train = fetch_20newsgroups(subset='train', categories=categories,
                                          shuffle=True, random_state=42,
                                          remove=('headers', 'footers', 'quotes'))
    newsgroups_test = fetch_20newsgroups(subset='test', categories=categories,
                                         shuffle=True, random_state=42,
                                         remove=('headers', 'footers', 'quotes'))

    X_train_raw, y_train = newsgroups_train.data, newsgroups_train.target
    X_test_raw, y_test = newsgroups_test.data, newsgroups_test.target

    print(f"Total training samples loaded: {len(X_train_raw)}")
    print(f"Total testing samples loaded: {len(X_test_raw)}")
    print(f"Target class names: {newsgroups_train.target_names}")

    print("\nExtracting features using CountVectorizer...")

    vectorizer = CountVectorizer()

    X_train_counts = vectorizer.fit_transform(X_train_raw)
    X_test_counts = vectorizer.transform(X_test_raw)

    print(f"Shape of training feature matrix: {X_train_counts.shape}")

    mnb_classifier = MultinomialNB()
    print("\nTraining Multinomial Naïve Bayes Classifier...")

    start_time = time.time()
    mnb_classifier.fit(X_train_counts, y_train)
    end_time = time.time()

    y_pred = mnb_classifier.predict(X_test_counts)

    accuracy = metrics.accuracy_score(y_test, y_pred)

    print("\n--- Model Evaluation Results ---")
    print(f"Training Time: {end_time - start_time:.4f} seconds")
    print(f"Accuracy Score: {accuracy:.4f}")

    print("\nClassification Report:")
    print(metrics.classification_report(y_test, y_pred, target_names=newsgroups_train.target_names))

    test_doc = ["The latest graphics drivers fixed the flickering problem on my screen."]
    test_doc_counts = vectorizer.transform(test_doc)
    predicted_category_index = mnb_classifier.predict(test_doc_counts)[0]
    predicted_category = newsgroups_train.target_names[predicted_category_index]

    print(f"\nExample Prediction for '{test_doc[0]}':")
    print(f"-> Predicted Category: {predicted_category}")


if __name__ == "__main__":
    np.random.seed(42)
    train_and_evaluate_multinomial_nb()
```

27. Write a Python program to train an SVM Classifier with different C values and compare the decision
boundaries visually

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_moons

def plot_decision_boundary(X, y, classifier, C_value, ax):

    x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
    y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                         np.arange(y_min, y_max, 0.02))

    Z = classifier.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    ax.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)

    ax.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, s=50, edgecolors='k')

    ax.scatter(classifier.support_vectors_[:, 0],
               classifier.support_vectors_[:, 1],
               s=150, facecolors='none', edgecolors='green', linewidth=2, label='Support Vectors')

    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
    ax.set_title(f'SVM Decision Boundary (C={C_value})')
    ax.legend()
    ax.grid(True)

def train_svm_with_different_c_and_visualize():

    print("--- Starting SVM C Value Comparison Visualization ---")

    X, y = make_moons(n_samples=200, noise=0.15, random_state=42)

    print(f"Generated {len(X)} samples with {X.shape[1]} features.")
    print(f"Number of samples in class 0: {np.sum(y == 0)}")
    print(f"Number of samples in class 1: {np.sum(y == 1)}")

    C_values = [0.1, 100.0]

    fig, axes = plt.subplots(1, 2, figsize=(18, 7))
    fig.suptitle('Comparison of SVM Decision Boundaries with Different C Values (RBF Kernel)', fontsize=16)

    for i, C_val in enumerate(C_values):

        print(f"\nTraining SVM Classifier with RBF Kernel (C={C_val}, gamma='scale')...")
        svm_classifier = svm.SVC(kernel='rbf', C=C_val, gamma='scale', random_state=42)
        svm_classifier.fit(X, y)

        print(f"C={C_val}: Number of Support Vectors: {len(svm_classifier.support_vectors_)}")
        print(f"C={C_val}: Training Accuracy: {svm_classifier.score(X, y):.4f}")

        plot_decision_boundary(X, y, svm_classifier, C_val, axes[i])

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
    print("\nVisualization complete. A plot window should have appeared showing two graphs.")

if __name__ == "__main__":
    np.random.seed(42)
    train_svm_with_different_c_and_visualize()
```

```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report

def compare_class_weighting_effect():

    print("--- Starting SVM Class Weighting Comparison on Imbalanced Data ---")

    X, y = make_classification(
        n_samples=1000,
        n_features=10,
        n_informative=5,
        n_redundant=0,
        n_classes=2,
        n_clusters_per_class=1,
        weights=[0.95, 0.05],
        flip_y=0,
        random_state=42
    )

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42, stratify=y
    )

    minority_count = np.sum(y_train == 1)
    majority_count = np.sum(y_train == 0)

    print(f"Training data summary:")
    print(f"  Majority Class (0) samples: {majority_count}")
    print(f"  Minority Class (1) samples: {minority_count}")
    print(f"  Imbalance Ratio: {majority_count / minority_count:.2f}:1")

    print("\n\n=== Model 1: SVM WITHOUT Class Weighting ===")

    svm_unweighted = SVC(kernel='rbf', random_state=42)
    svm_unweighted.fit(X_train, y_train)
    y_pred_unweighted = svm_unweighted.predict(X_test)

    print("Classification Report (Unweighted):")

    print(classification_report(y_test, y_pred_unweighted))

    print("\n\n=== Model 2: SVM WITH Class Weighting (class_weight='balanced') ===")

    svm_weighted = SVC(kernel='rbf', class_weight='balanced', random_state=42)
    svm_weighted.fit(X_train, y_train)
    y_pred_weighted = svm_weighted.predict(X_test)

    print("Classification Report (Weighted):")

    print(classification_report(y_test, y_pred_weighted))

    print("\n=======================================================")
    print("                 Summary of Minority Class (Class 1) Metrics")
    print("=======================================================")

    report_unweighted = classification_report(y_test, y_pred_unweighted, output_dict=True)
    report_weighted = classification_report(y_test, y_pred_weighted, output_dict=True)

    unweighted_recall = report_unweighted['1']['recall']
    weighted_recall = report_weighted['1']['recall']

    print(f"Minority Class Recall (Unweighted): {unweighted_recall:.4f}")
    print(f"Minority Class Recall (Weighted):   {weighted_recall:.4f}")

    if weighted_recall > unweighted_recall:
        print("\nConclusion: The 'balanced' class weighting successfully improved the model's ability to identify the minority class (higher recall)!")
    else:
        print("\nConclusion: Class weighting did not significantly improve minority class recall in this specific instance.")
    print("=======================================================")


if __name__ == "__main__":
    np.random.seed(42)
    compare_class_weighting_effect()
```

37. Write a Python program to train an SVM Classifier using Linear, Polynomial, and RBF kernels on the Breast

```python
import time
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_breast_cancer

print("--- 1. Loading and Preparing Data ---")
data = load_breast_cancer()
X = data.data
y = data.target

print(f"Dataset shape: {X.shape}")
print(f"Feature Names: {data.feature_names.size} features")
print(f"Target Classes: {data.target_names.size} classes\n")

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

def train_and_evaluate_svm(kernel_type, X_train, X_test, y_train, y_test, **kwargs):
    """Trains an SVM with a specified kernel and returns accuracy and training time."""
    print(f"--- Training SVM with {kernel_type.upper()} Kernel ---")

    if kernel_type == 'poly':

        model = SVC(kernel=kernel_type, C=1.0, degree=3, random_state=42, **kwargs)
    else:

        model = SVC(kernel=kernel_type, C=1.0, random_state=42, **kwargs)

    start_time = time.time()
    model.fit(X_train, y_train)
    training_time = time.time() - start_time

    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)

    print(f"  Accuracy: {accuracy:.4f}")
    print(f"  Training Time: {training_time:.4f} seconds\n")
    return accuracy, training_time


results = {}

acc_linear, time_linear = train_and_evaluate_svm(
    'linear', X_train_scaled, X_test_scaled, y_train, y_test
)
results['Linear'] = {'Accuracy': acc_linear, 'Time': time_linear}

acc_poly, time_poly = train_and_evaluate_svm(
    'poly', X_train_scaled, X_test_scaled, y_train, y_test
)
results['Polynomial'] = {'Accuracy': acc_poly, 'Time': time_poly}

acc_rbf, time_rbf = train_and_evaluate_svm(
    'rbf', X_train_scaled, X_test_scaled, y_train, y_test
)
results['RBF'] = {'Accuracy': acc_rbf, 'Time': time_rbf}

print("\n" + "="*60)
print("     SVM CLASSIFIER KERNEL PERFORMANCE COMPARISON")
print("="*60)

sorted_results = sorted(results.items(), key=lambda item: item[1]['Accuracy'], reverse=True)

for kernel, metrics in sorted_results:
    print(f"Kernel: {kernel:<15} | Accuracy: {metrics['Accuracy']:.4f} | Training Time: {metrics['Time']:.4f}s")

print("="*60)

best_kernel = sorted_results[0][0]
best_accuracy = sorted_results[0][1]['Accuracy']

print(f"\nCONCLUSION: The {best_kernel} kernel achieved the highest accuracy of {best_accuracy:.4f} on this dataset.")
```

38. Write a Python program to train an SVM Classifier using Stratified K-Fold Cross-Validation and compute the

```python
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_breast_cancer

print("--- 1. Loading and Preparing Data ---")
data = load_breast_cancer()
X = data.data
y = data.target

print(f"Dataset shape: {X.shape}")
print(f"Number of target classes: {len(np.unique(y))}")
print(f"Class distribution: {np.bincount(y)}\n")

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

N_SPLITS = 10
print(f"--- 2. Setting up {N_SPLITS}-Fold Stratified K-Fold ---")

skf = StratifiedKFold(n_splits=N_SPLITS, shuffle=True, random_state=42)

fold_accuracies = []
fold_counter = 1

print("--- 3. Running Cross-Validation ---")
model = SVC(kernel='rbf', C=1.0, random_state=42)

for train_index, val_index in skf.split(X_scaled, y):

    X_train_fold, X_val_fold = X_scaled[train_index], X_scaled[val_index]
    y_train_fold, y_val_fold = y[train_index], y[val_index]

    model.fit(X_train_fold, y_train_fold)

    y_pred_fold = model.predict(X_val_fold)

    accuracy = accuracy_score(y_val_fold, y_pred_fold)
    fold_accuracies.append(accuracy)

    print(f"Fold {fold_counter}/{N_SPLITS} Accuracy: {accuracy:.4f}")
    fold_counter += 1

average_accuracy = np.mean(fold_accuracies)
std_dev_accuracy = np.std(fold_accuracies)

print("\n" + "="*50)
print("   STRATIFIED K-FOLD CROSS-VALIDATION RESULTS")
print("="*50)
print(f"Total Folds: {N_SPLITS}")
print(f"Individual Fold Accuracies: {np.round(fold_accuracies, 4)}")
print(f"Mean Accuracy: {average_accuracy:.4f}")
print(f"Standard Deviation: {std_dev_accuracy:.4f}")
print("="*50)
```

39. Write a Python program to train a Naïve Bayes classifier using different prior probabilities and compare
performance

```python
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_breast_cancer

print("--- 1. Loading and Preparing Data ---")
data = load_breast_cancer()
X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

class_counts = np.bincount(y_train)
actual_priors = class_counts / len(y_train)
print(f"Actual (Learned) Priors for Training Data (Class 0, Class 1): {actual_priors.round(4)}")

uniform_priors = [0.5, 0.5]

biased_priors = [0.9, 0.1]


results = {}

print("\n--- 3.1 Training GNB with Learned Priors (Default) ---")
nb_default = GaussianNB()
nb_default.fit(X_train_scaled, y_train)
y_pred_default = nb_default.predict(X_test_scaled)
acc_default = accuracy_score(y_test, y_pred_default)
results['Learned (Actual)'] = {'Priors': actual_priors.round(4), 'Accuracy': acc_default}
print(f"Accuracy: {acc_default:.4f}")

print("\n--- 3.2 Training GNB with Uniform Priors [0.5, 0.5] ---")
nb_uniform = GaussianNB(priors=uniform_priors)
nb_uniform.fit(X_train_scaled, y_train)
y_pred_uniform = nb_uniform.predict(X_test_scaled)
acc_uniform = accuracy_score(y_test, y_pred_uniform)
results['Uniform [0.5, 0.5]'] = {'Priors': uniform_priors, 'Accuracy': acc_uniform}
print(f"Accuracy: {acc_uniform:.4f}")

print("\n--- 3.3 Training GNB with Biased Priors [0.9, 0.1] ---")
nb_biased = GaussianNB(priors=biased_priors)
nb_biased.fit(X_train_scaled, y_train)
y_pred_biased = nb_biased.predict(X_test_scaled)
acc_biased = accuracy_score(y_test, y_pred_biased)
results['Biased [0.9, 0.1]'] = {'Priors': biased_priors, 'Accuracy': acc_biased}
print(f"Accuracy: {acc_biased:.4f}")

print("\n" + "="*70)
print("     GAUSSIAN NAÏVE BAYES CLASSIFIER PRIORS COMPARISON")
print("="*70)

print(f"{'Strategy':<20} | {'Priors (Class 0, Class 1)':<30} | {'Accuracy':<10}")
print("-" * 70)
for strategy, metrics in results.items():
    priors_str = f"[{metrics['Priors'][0]:.4f}, {metrics['Priors'][1]:.4f}]"
    print(f"{strategy:<20} | {priors_str:<30} | {metrics['Accuracy']:.4f}")

print("="*70)
print("\nNote: The 'Learned' model typically performs best as its priors reflect the actual test data distribution.")
```

40. Write a Python program to perform Recursive Feature Elimination (RFE) before training an SVM Classifier and
compare accuracy

```python
import time
import numpy as np
from sklearn.svm import SVC
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_breast_cancer

print("--- 1. Loading and Preparing Data ---")
data = load_breast_cancer()
X = data.data
y = data.target

TOTAL_FEATURES = X.shape[1]

N_FEATURES_TO_SELECT = TOTAL_FEATURES // 2

print(f"Total features available: {TOTAL_FEATURES}")
print(f"Number of features to select using RFE: {N_FEATURES_TO_SELECT}\n")

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("--- 2. Training Baseline SVM (All Features) ---")
start_time = time.time()
svm_baseline = SVC(kernel='linear', C=1.0, random_state=42)
svm_baseline.fit(X_train_scaled, y_train)
time_baseline = time.time() - start_time

y_pred_baseline = svm_baseline.predict(X_test_scaled)
acc_baseline = accuracy_score(y_test, y_pred_baseline)

print(f"Baseline Accuracy (All Features): {acc_baseline:.4f}")
print(f"Training Time: {time_baseline:.4f} seconds\n")

print("--- 3. Performing Recursive Feature Elimination (RFE) ---")

estimator = SVC(kernel='linear', C=1.0, random_state=42)

rfe = RFE(estimator=estimator, n_features_to_select=N_FEATURES_TO_SELECT, step=1)

rfe.fit(X_train_scaled, y_train)

selected_features_indices = np.where(rfe.support_)[0]
selected_feature_names = data.feature_names[selected_features_indices]

print(f"Selected feature indices: {selected_features_indices}")
print(f"Selected feature names ({N_FEATURES_TO_SELECT}):")
for name in selected_feature_names:
    print(f"  - {name}")

X_train_rfe = rfe.transform(X_train_scaled)
X_test_rfe = rfe.transform(X_test_scaled)

print("\n--- 4. Training RFE-Optimized SVM (Selected Features) ---")
start_time = time.time()

svm_rfe = SVC(kernel='linear', C=1.0, random_state=42)
svm_rfe.fit(X_train_rfe, y_train)
time_rfe = time.time() - start_time

y_pred_rfe = svm_rfe.predict(X_test_rfe)
acc_rfe = accuracy_score(y_test, y_pred_rfe)

print(f"RFE-Optimized Accuracy ({N_FEATURES_TO_SELECT} features): {acc_rfe:.4f}")
print(f"Training Time: {time_rfe:.4f} seconds\n")

print("\n" + "="*70)
print("   SVM CLASSIFIER PERFORMANCE COMPARISON (RFE vs. Baseline)")
print("="*70)
print(f"{'Model Type':<30} | {'Features':<10} | {'Accuracy':<10} | {'Train Time (s)':<15}")
print("-" * 70)
print(f"{'Baseline (All Features)':<30} | {TOTAL_FEATURES:<10} | {acc_baseline:.4f} | {time_baseline:.4f}")
print(f"{'RFE-Optimized':<30} | {N_FEATURES_TO_SELECT:<10} | {acc_rfe:.4f} | {time_rfe:.4f}")
print("="*70)
```

41. Write a Python program to train an SVM Classifier and evaluate its performance using Precision, Recall, and

```python
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_breast_cancer

print("--- 1. Loading and Preparing Data ---")
data = load_breast_cancer()
X = data.data
y = data.target
target_names = data.target_names

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("\n--- 2. Training the SVM Classifier (RBF Kernel) ---")
svm_model = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42)
svm_model.fit(X_train_scaled, y_train)

y_pred = svm_model.predict(X_test_scaled)

print("\n--- 3. Comprehensive Performance Evaluation ---")

acc = accuracy_score(y_test, y_pred)

report = classification_report(y_test, y_pred, target_names=target_names)

print(f"Overall Accuracy: {acc:.4f}\n")
print("Detailed Classification Report:")
print("-" * 50)
print(report)
print("-" * 50)

print("Metric Interpretation:")
print(f"Precision: The accuracy of positive predictions (when the model predicts {target_names[1]}, how often is it correct?).")
print(f"Recall: The fraction of all positive samples that were correctly predicted as positive (how many actual {target_names[1]} cases did the model catch?).")
print("F1-Score: The harmonic mean of Precision and Recall (a balance between the two).")
```

45.  Write a Python program to train a Naïve Bayes classifier and evaluate its performance using the ROC-AUC

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, roc_auc_score, accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_breast_cancer

print("--- 1. Loading and Preparing Data ---")
data = load_breast_cancer()
X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("\n--- 2. Training Gaussian Naïve Bayes Classifier ---")
nb_model = GaussianNB()
nb_model.fit(X_train_scaled, y_train)

y_probs = nb_model.predict_proba(X_test_scaled)[:, 1]

y_pred_class = nb_model.predict(X_test_scaled)
acc = accuracy_score(y_test, y_pred_class)

print("\n--- 3. Performance Evaluation and Visualization ---")

roc_auc = roc_auc_score(y_test, y_probs)

fpr, tpr, thresholds = roc_curve(y_test, y_probs)

print(f"Overall Accuracy (Threshold 0.5): {acc:.4f}")
print(f"ROC-AUC Score: {roc_auc:.4f}")

plt.figure(figsize=(8, 6))

plt.plot(
    fpr, tpr,
    color='darkorange',
    lw=2,
    label=f'ROC curve (AUC = {roc_auc:.4f})'
)

plt.plot(
    [0, 1], [0, 1],
    color='navy',
    lw=2,
    linestyle='--',
    label='Random Classifier (AUC = 0.50)'
)

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (FPR)', fontsize=14)
plt.ylabel('True Positive Rate (TPR) / Recall', fontsize=14)
plt.title('Receiver Operating Characteristic (ROC) Curve for Naïve Bayes', fontsize=16)
plt.legend(loc="lower right")
plt.grid(True)
plt.tight_layout()

plt.show()

print("\nVisualization Complete. The ROC curve plot has been generated.")
```

46.  Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_recall_curve, average_precision_score, auc
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_breast_cancer

print("--- 1. Loading and Preparing Data ---")
data = load_breast_cancer()
X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("\n--- 2. Training SVM Classifier (RBF Kernel, with probability estimates) ---")

svm_model = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42, probability=True)
svm_model.fit(X_train_scaled, y_train)

y_scores = svm_model.predict_proba(X_test_scaled)[:, 1]

print("\n--- 3. Computing Precision-Recall Curve Metrics ---")
precision, recall, thresholds = precision_recall_curve(y_test, y_scores)

average_precision = average_precision_score(y_test, y_scores)
pr_auc = auc(recall, precision)

print(f"Average Precision (AP) Score: {average_precision:.4f}")
print(f"Area Under Precision-Recall Curve (AUC-PR): {pr_auc:.4f}")

plt.figure(figsize=(8, 6))

plt.plot(
    recall, precision,
    color='blue',
    lw=2,
    label=f'Precision-Recall curve (AP = {average_precision:.4f})'
)

no_skill = len(y_test[y_test==1]) / len(y_test)
plt.plot(
    [0, 1], [no_skill, no_skill],
    linestyle='--',
    color='red',
    label=f'No-Skill Line (P = {no_skill:.2f})'
)

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Recall', fontsize=14)
plt.ylabel('Precision', fontsize=14)
plt.title('SVM Classifier: Precision-Recall Curve', fontsize=16)
plt.legend(loc="lower left")
plt.grid(True)
plt.tight_layout()

plt.show()

print("\nVisualization Complete. The Precision-Recall curve plot has been generated.")
```

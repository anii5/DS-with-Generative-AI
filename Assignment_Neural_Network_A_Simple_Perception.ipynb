1. What is Deep Learning? Briefly describe how it evolved and how it differs from traditional machine learning.
  - Deep Learning is a specialized subset of Machine Learning (ML) that mimics the way the human brain processes data through layers of artificial neurons. While traditional ML often requires humans to identify and label features (like telling a computer that "whiskers" and "pointed ears" define a cat), Deep Learning models can automatically discover these features on their own.
  - Deep Learning vs. Traditional Machine Learning
The easiest way to distinguish the two is by looking at how they learn and how they scale.

1. Feature Engineering
Machine Learning: Requires "Feature Extraction." A human expert must manually select which variables (e.g., pixel intensity, edges, shapes) are important for the model to analyze.

Deep Learning: Performs "Feature Learning." It passes raw data (like an image) through layers that progressively learn complex patterns, from simple edges in the first layer to full faces in the final layers.
+1

2. Data Dependency
Machine Learning: Performs well on smaller datasets. Adding more data beyond a certain point often yields diminishing returns.

Deep Learning: It is "data-hungry." Its performance continues to improve as you feed it more data, which is why it thrives in the era of Big Data.

3. Hardware Requirements
Machine Learning: Can usually run on a standard CPU.

Deep Learning: Requires massive computational power, specifically GPUs (Graphics Processing Units) or TPUs, to handle the millions of matrix multiplications happening simultaneously.

2.  Explain the basic architecture and functioning of a Perceptron. What are its limitations?
  - Limitation of Perceptron Model
The output of a perceptron can only be a binary number (0 or 1) due to the hard-edge transfer function. It can only be used to classify the linearly separable sets of input vectors. If the input vectors are non-linear, it is not easy to classify them correctly.
  - The Architecture of a PerceptronThink of a Perceptron as a mathematical decision-making unit. It takes several binary or numerical inputs and produces a single binary output (0 or 1).
      - Key Components:
          - Inputs ($x_1, x_2, ... x_n$): These are the features or data points you feed into the model.
          - Weights ($w_1, w_2, ... w_n$): Each input has a weight that represents its relative importance. If a weight is high, that specific input has a larger impact on the output.
          - Bias ($b$): A constant value added to the weighted sum. It allows the model to shift the activation function up or down, providing flexibility in decision-making.
          - Summation Function: The Perceptron calculates the weighted sum of all inputs plus the bias.
          - Activation Function (Step Function): This "squashes" the sum into a binary output. If the sum is above a certain threshold, the output is 1; otherwise, it is 0.

3. Describe the purpose of activation function in neural networks. Compare Sigmoid, ReLU, and Tanh functions.
  - The activation function is the "gatekeeper" of a neural network. Its primary purpose is to introduce non-linearity into the model.
Without an activation function, no matter how many layers you stack, the entire neural network would behave like a single linear regression model. Since real-world data (images, speech, stock market trends) is rarely linear, activation functions allow the network to learn complex patterns and solve non-linear problems like the XOR issue mentioned earlier.

    - 1. Sigmoid Function : The Sigmoid function scales any input into a value between 0 and 1.
    - 2. Tanh Function : Tanh is essentially a scaled version of Sigmoid that ranges from -1 to 1.
    - 3. ReLU (Rectified Linear Unit) : ReLU is the most widely used activation function in modern Deep Learning. It outputs the input directly if it is positive; otherwise, it outputs zero.

4. What is the difference between Loss function and Cost function in neural networks? Provide examples.
  - While the terms Loss Function and Cost Function are often used interchangeably in casual conversation, they have a specific technical distinction in the context of training a neural network.The simplest way to remember the difference is: Loss is for a single data point, while Cost is for the entire dataset.
      - Loss Function (The "Per-Sample" Error)The Loss Function measures the discrepancy between the predicted output ($\hat{y}$) and the actual target ($y$) for one single training example. It tells the network how "wrong" it was on a specific instance (e.g., one image of a cat).
      - Cost Function (The "Global" Error)
The Cost Function is the average (or sum) of all the individual loss functions over the entire training set. It represents the total "price" the model pays for its current set of weights and biases across all data points.
The goal of optimization algorithms (like Gradient Descent) is to minimize this Cost Function.

5. What is the role of optimizers in neural networks? Compare Gradient Descent, Adam, and RMSprop.
  - If the Cost Function is the map that tells a neural network how far it is from the goal, the Optimizer is the driver that decides which direction to turn the steering wheel and how fast to press the gas pedal.
The role of an optimizer is to update the weights and biases of the network to minimize the Cost Function. It uses the gradients (slopes) calculated during backpropagation to navigate the "landscape" of the cost function toward the lowest point, known as the Global Minimum.
      - Gradient Descent (The Foundation) : The most basic optimizer. It updates weights by taking a step in the opposite direction of the gradient.
      - RMSprop (Root Mean Squared Propagation): RMSprop was designed to solve the problem of gradients oscillating wildly or vanishing.
      - Adam (Adaptive Moment Estimation)
Adam is currently the "gold standard" and the most widely used optimizer in deep learning. It essentially combines the best parts of RMSprop and Momentum.

6. Write a Python program to implement a single-layer perceptron from scratch using NumPy to solve the logical AND gate.

```python
import numpy as np

def unitStep(v):
    if v >= 0:
        return 1
    else:
        return 0

def perceptronModel(x, w, b):
    v = np.dot(w, x) + b
    y = unitStep(v)
    return y

def AND_logicFunction(x):
    w = np.array([1, 1])
    b = -1.5
    return perceptronModel(x, w, b)

test1 = np.array([0, 1])
test2 = np.array([1, 1])
test3 = np.array([0, 0])
test4 = np.array([1, 0])

print("AND({}, {}) = {}".format(0, 1, AND_logicFunction(test1)))
print("AND({}, {}) = {}".format(1, 1, AND_logicFunction(test2)))
print("AND({}, {}) = {}".format(0, 0, AND_logicFunction(test3)))
print("AND({}, {}) = {}".format(1, 0, AND_logicFunction(test4)))
```

7. Implement and visualize Sigmoid, ReLU, and Tanh activation functions using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

x = np.linspace(-10, 10, 100)

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(x, sigmoid(x), 'b', linewidth=2)
plt.title("Sigmoid Function")
plt.grid(True)
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)

plt.subplot(2, 2, 2)
plt.plot(x, tanh(x), 'r', linewidth=2)
plt.title("Tanh Function")
plt.grid(True)
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)

plt.subplot(2, 2, 3)
plt.plot(x, relu(x), 'g', linewidth=2)
plt.title("ReLU Function")
plt.grid(True)
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)

plt.subplot(2, 2, 4)
plt.plot(x, sigmoid(x), label='Sigmoid')
plt.plot(x, tanh(x), label='Tanh')
plt.plot(x, relu(x), label='ReLU')
plt.title("Comparison")
plt.legend()
plt.grid(True)
plt.ylim(-1.5, 2)

plt.tight_layout()
plt.show()
```

8.  Use Keras to build and train a simple multilayer neural network on the MNIST digits dataset. Print the training accuracy.

```python
import tensorflow as tf
from tensorflow.keras import layers, models

mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train, x_test = x_train / 255.0, x_test / 255.0

model = models.Sequential([

    layers.Flatten(input_shape=(28, 28)),

    layers.Dense(128, activation='relu'),

    layers.Dropout(0.2),

    layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

print("Starting training...")
history = model.fit(x_train, y_train, epochs=5, verbose=1)

train_loss, train_acc = model.evaluate(x_train, y_train, verbose=0)
print(f"\nFinal Training Accuracy: {train_acc * 100:.2f}%")
```

9. Visualize the loss and accuracy curves for a neural network model trained on the Fashion MNIST dataset. Interpret the training behavior.

```python
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt

fashion_mnist = tf.keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

train_images = train_images / 255.0
test_images = test_images / 255.0

model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(256, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(train_images, train_labels, epochs=15,
                    validation_split=0.2, verbose=1)

acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs_range = range(1, len(acc) + 1)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.grid(True)

plt.tight_layout()
plt.show()
```

10. You are working on a project for a bank that wants to automatically detect fraudulent transactions. The dataset is large, imbalanced, and contains structured features like transaction amount, merchant ID, and customer location. The goal is to classify each transaction as fraudulent or legitimate.

Explain your real-time data science workflow:
  - How would you design a deep learning model (perceptron or multilayer NN)?
  - Which activation function and loss function would you use, and why?
  - How would you train and evaluate the model, considering class imbalance?
  - Which optimizer would be suitable, and how would you prevent overfitting?

```python
import tensorflow as tf
from tensorflow.keras import layers, models

def build_fraud_model(input_dim):
    model = models.Sequential([
        layers.Dense(128, activation='relu', input_shape=(input_dim,)),
        layers.BatchNormalization(),
        layers.Dropout(0.3),

        layers.Dense(64, activation='relu'),
        layers.Dropout(0.2),

        layers.Dense(1, activation='sigmoid')
    ])

    model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=[tf.keras.metrics.Recall(), tf.keras.metrics.Precision()])
    return model
```
